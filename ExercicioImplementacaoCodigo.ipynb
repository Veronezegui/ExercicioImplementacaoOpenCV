```python
# Alunos: Guilherme da Silva Veroneze, Hudson Loyo Guimar√£es e Yago Henrique Nascimento Moreira Rosestolato

import numpy as np
import pandas as pd
import cv2 as cv 
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
import scipy

# !wget --no-check-certificate \
#     https://raw.githubusercontent.com/computationalcore/introduction-to-opencv/master/utils/common.py \
#     -O common.py
```

```python
# Criando uma lista de urls de imagens de pessoas sorrindo que vamos utilizar posteriormente:
urls = ["face.jpg"]

# Lendo e mostrando cada imagem da lista em BGR e RGB:
for url in urls:
  image_s = io.imread(url) 
  image_s2 = cv.cvtColor(image_s, cv.COLOR_BGR2RGB)
  
  print('\n')

  image_s2_PB = cv.cvtColor(image_s2, cv.COLOR_BGR2GRAY)
  cv2_imshow(image_s2_PB)


# Carregando o modelo pr√©-treinado de detec√ß√£o de sorrisos:
smile_cascade = cv.CascadeClassifier(cv.data.haarcascades + 'haarcascade_frontalface_default.xml')
smiles = smile_cascade.detectMultiScale(image_s2_PB, 1.3, 20)
for (x,y,w,h) in smiles:
# Calculando as coordenadas do ponto m√©dio da face e desenhando o ret√¢ngulo
     topo_x = ((x+w)/2)
     topo_y = ((y+h)/2)
     raio = ((w)/2)
     cv.rectangle(image_s2,(x,y),(x+w,y+h),(0,0,255),8)
# Desenhando o c√≠rculo no ponto m√©dio da face, com a cor verde e espessura 8
     cv.circle(image_s2, (int(topo_x),int(topo_y)), (int(raio)), (100,255,0), 8)




cv2_imshow(image_s2)

print('\n')

```
